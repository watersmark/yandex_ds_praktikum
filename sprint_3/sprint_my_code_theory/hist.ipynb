## Рассмотрим гистограммы 

### В примере ниже, мы вызываем функцию hist(bins=3,range=()) для создания гистограммы, данную гистограмму мы делим на три части методом bins=3, т.е для первого массива   'length': [1.5, 0.5, 1.2, 0.9, 3], мы получим шаг (3 - 0.5) / 3 = 0.83, и начиная с 0.5 + 0.83 = 1.33, мы разобъём наш набор данных на три части, где по оси (ox) у нас идут шаги, а по оси (oy) кол-во элементов в промежутке 

```python
import pandas as pd

df = pd.DataFrame({
    'length': [1.5, 0.5, 1.2, 0.9, 3],
    'width': [0.7, 0.2, 0.15, 0.2, 1.1]
    }, index=['pig', 'rabbit', 'duck', 'chicken', 'horse'])
hists = df.hist(bins=3)
```

### Так же надо отметить, что значение слева всегда считается входящим, а справа нет, т.е [0.5, 1.83), нона последнем отрезке, левый и правый конец оба включены [2, 3]

```python
df = pd.Series([1, 2, 3])
df.hist(bins=2)
```

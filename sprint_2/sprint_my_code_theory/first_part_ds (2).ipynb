### Чтобы сгрупирровать данные по стобцу используем метод groupby('<col_name>'), надо помнить, что при группировке, столбцы у которых значения None или NaN. При группировке по данному столбцу, в данном столбце все нулевые значения удаляются. Так же общий тип значений, если все нулевые определяется, как тип object, если два значения(int и str), то перевес будет к типу object

```python
df.groupby('children')['children'].count()
```

### Чтобы при работе со столбцами не вызывать множество раз повторный код, используем метод agg(), это словарь, где можно указать столбец и методы, которые будут к нему применены. После аггрегирования обращение происходит по двойному индексу

```python
df.agg({'children' : ['sum', 'count']})
temp_df = df.groupby('children').agg({'dob_years': ['sum', 'count']})
temp_df['dob_years']
```

### Для преобразования одного типа данных в другой используем pd.to_numeric(), где в скобках указываем преобразуемые данные, а так же, что желать при ошибке: 
### errors= raise' — дефолтное поведение: при встрече с некорректным значением выдаётся ошибка, операция перевода в числа прерывается;
### errors= coerce' — некорректные значения принудительно заменяются на NaN;
### errors= ignore' — некорректные значения игнорируются, но остаются.

```python
pd.to_numeric(df['days_employed'], errors='coerce')
```

### Для выделения основы слова(убираем окончание) можно использовать стемминг, в python уже есть библиотека, реализующая стемминг nltk

```python
from nltk.stem import SnowballStemmer
russian_stemmer = SnowballStemmer('russian')

for reason_buy in df['purpose']:
    for word in reason_buy.split():
        stem_word = russian_stemmer.stem(word)
        if stem_word == 'покупк':
            print(reason_buy)
            break
```

### Для лемматизации слов используем библиотеку pymystem3. Формы записи в словаре русских слов:
для существительных — именительный падеж, единственное число;
для прилагательных — именительный падеж, единственное число, мужской род;
для глаголов, причастий, деепричастий — глагол в инфинитиве несовершенного вида.
Для многих случаев это гораздо лучше основы слова, которую возвращает NLTK.

### Библиотека работает не быстро

```python
import pandas as pd
from pymystem3 import Mystem

df = pd.read_csv('data.csv')
russ_stem = Mystem()


for reason_word in df['purpose']:
    if 'покупка' in russ_stem.lemmatize(reason_word):
        print(reason_word)
```

### Для применения метода ко всем столбцам или строкам или отдельным столбцам используется apply()
Когда метод apply() оперирует данными из нескольких столбцов, его вызывают ко всему датафрейму. В таком случае он принимает не только название функции, но и параметр axis: со значением 1, чтобы применить метод ко всем строкам датафрейма, и со значением 0 — ко всем столбцам.

### Можно вызывать, как для одного столбца, так и для нескольких с указанием axis


```python
def sum_param(arg):
    try:
        days_on_work = arg['days_employed']
        years_of_people = arg['dob_years']
        return days_on_work * years_of_people
    except:
        return 0
    
df['new_col'] = df[['days_employed', 'dob_years']].apply(sum_param, axis=1)
df.head()
```
